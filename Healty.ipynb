### Data Preprocessing

Checking data types of all features

```python
df_train.info()
```

Observing that certain columns do not consist of numerical values, it becomes important to determine the number of unique values in each column. This will aid in selecting an appropriate encoding method for categorical columns, considering that most machine learning models operate effectively with numerical data.

```python
pd.Series({c: df_train[c].unique().size for c in df_train})
```

```python
df_train = pd.get_dummies(df_train, columns = ['Smoker?','Living in?','Food preference','Any heriditary condition?','Physical activity'])
df_train.head()
```

Now, checking for missing values using heatmap

```python
sns.heatmap(df_train.isnull(), cmap='Blues', cbar=False, yticklabels=False, xticklabels=df_train.columns);
```

We can clearly see from the above heatmap that many columns contain missing values. Applying Iterative Imputer to all feature colums to handle missing values.

```python
feature_cols = [col for col in df_train.columns if col not in ['Healthy']]
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

itr_imputer = IterativeImputer()

df_train[feature_cols] = itr_imputer.fit_transform(df_train[feature_cols])
sns.heatmap(df_train.isnull(), cmap='Blues', cbar=False, yticklabels=False, xticklabels=df_train.columns)
```

Now, there are no missing values left

Storing Feature Matrix In X and Response(Target) in Vector y

```python
X = df_train.drop('Healthy', axis=1)
y = df_train['Healthy']
```

Splitting the dataset into the Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

Now let's perform Machine Learning Classification. We'll be predicting the target variable based on the given features.

### Model Selection

We will use several popular models and see how they perform on our dataset. At the end we will choose the best performing model.

Importing models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier, XGBRFClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

Creating a Predefined function to assess the performance of a model

```python
def performance(model):
    
    model.fit(X_train, y_train)
    y_train_hat = model.predict(X_train)
    y_test_hat = model.predict(X_test)

    print(model)
    print('Train performance')
    print('-------------------------------------------------------')
    print(classification_report(y_train, y_train_hat))

    print('Test performance')
    print('-------------------------------------------------------')
    print(classification_report(y_test, y_test_hat))

    print('Roc_auc score')
    print('-------------------------------------------------------')
    print(roc_auc_score(y_test, y_test_hat))
    print('')

    print('Confusion matrix')
    print('-------------------------------------------------------')
    print(confusion_matrix(y_test, y_test_hat))
```

Next, we will train the models individually and evaluate their performance on our dataset.

```python
# Logistic Regression
lr = LogisticRegression()
performance(lr)
```

```python
# Naive Bayes: GaussianNB
gnb = GaussianNB()
performance(gnb)
```

```python
# Support Vector Classifier
svc = SVC()
performance(svc)
```

```python
# Decision Tree Classifier
dtc = DecisionTreeClassifier()
performance(dtc)
```

```python
# Random Forest Classifier
rfc = RandomForestClassifier()
performance(rfc)
```

```python
# AdaBoost Classifier
abc = AdaBoostClassifier()
performance(abc)
```

```python
# XGBoost Classifier
xgbc = XGBClassifier()
performance(xgbc)
```

```python
# XGBRF Classifier
xgbrf = XGBRFClassifier()
performance(xgbrf)
```

```python
# K-Nearest Neighbors Classifier
knnc = KNeighborsClassifier()
performance(knnc)
```
